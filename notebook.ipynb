```
import base64import jsonimport sysimport osimport githubfrom github import Githubimport hashlibfrom flask import Flask, redirect, request
```

```
gh = Github(os.environ['GITHUB_TOKEN'])user = gh.get_user()
```

```
app = Flask("colaburl")
```

```
@app.route("/", methods=['GET'])def main():  assert len({'py', 'py64', 'ipynb'} & request.args.keys()) == 1  if 'ipynb' in request.args:    content = request.args['ipynb']  else:    if 'py' in request.args:      content = request.args['py']    elif 'py64' in request.args:      content = base64.b64decode(request.args['py64'])    cells = content.split('\n\n')    content = json.dumps({      "nbformat": 4,      "nbformat_minor": 0,      "metadata": {},      "cells": [        {          "cell_type": "code",          "source": cell.split('\n'),          "metadata": {"id": str(i)}        } for i, cell in enumerate(cells)      ]    })  name = request.args.get('name', 'notebook.ipynb')  assert '/' not in name  assert name.endswith('.ipynb')
```

```
  id = hashlib.sha256(json.dumps({'name': name, 'content': content}).encode()).hexdigest()  repo = None  try:    repo = user.create_repo(id)  except github.GithubException as ex:    if ex.status != 422:      raise
```

```
  if repo:    repo.create_file(name, 'autocommit', content)
```

```
  return redirect(f'https://colab.research.google.com/github/colaburl/{id}/blob/main/{name}')
```
